```python
# Import dependencies
import pandas as pd
import numpy as np
import requests
```

```python
sales = pd.read_csv("vgsales.csv")
sales.head()
```

```python
sales = sales.rename(columns={
    'Rank':'uniqueid',
    'Name':'name',
    'Platform':'platform',
    'Year':'year',
    'Genre':'genre',
    'Publisher':'publisher',
    'NA_Sales':'nasales',
    'EU_Sales':'eusales',
    'JP_Sales':'jpsales',
    'Other_Sales':'othersales',
    'Global_Sales':'globalsales'
})
sales.head()
```

```python
sales = sales.drop(['platform','year','genre','publisher'], axis=1)
```

```python
sales.head()
```

```python
sales.to_csv('video_game_sales.csv',encoding='utf8',index=False)
```

```python
info = pd.read_csv("video_game_info.csv")
info.head()
```

```python
info.info
```

```python
#Merging the two datasets
merged = info.merge(sales,on='uniqueid',how='left')
merged
```

```python
#Drop any unwanted columns
merged = merged.drop(['name_y'],axis=1)
merged = merged.rename(columns={
    'name_x':'name'})
merged
```

```python
#Convert to a csv file
merged.to_csv('video_game_cleaned.csv',encoding='utf8',index=False)
```

```python
#Determining Total Sales by Genre
genre_sales = pd.DataFrame(merged.groupby(["genre"])["globalsales"].sum().sort_values(ascending=False))
genre_sales
```

```python
#Determining Average Sales by Genre
avg_genre_sales = pd.DataFrame(merged.groupby(["genre"])["globalsales"].mean().sort_values(ascending=False))
avg_genre_sales
```

```python
#Determining Total Sales by Game
game_sales = pd.DataFrame(merged.groupby(["name","genre"])["globalsales"].sum().sort_values(ascending=False))
game_sales.head(10)
```

```python
#Determining Average Sales by Game
avg_game_sales = pd.DataFrame(merged.groupby(["name"])["globalsales"].mean().sort_values(ascending=False))
avg_game_sales
```
